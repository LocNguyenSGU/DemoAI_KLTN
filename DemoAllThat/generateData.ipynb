```python
import pandas as pd
import numpy as np

def generate_student_log_data(n_samples=5000, random_state=None):
    rng = np.random.default_rng(random_state)

    data = []

    for _ in range(n_samples):
        student = {}

        for i in range(1, 8):  # M·ª•c 1.1 ‚Üí 1.7
            # Gi·∫£ l·∫≠p m·ª©c ƒë·ªô t∆∞∆°ng t√°c c·ªßa sinh vi√™n
            pdf_views = rng.integers(0, 6)
            video_views = rng.integers(0, 6)

            student[f"pdf_views_1_{i}"] = pdf_views
            student[f"video_views_1_{i}"] = video_views

            # Quiz ƒëi·ªÉm d·ªÖ / TB / kh√≥
            student[f"quiz_avg_easy_1_{i}"] = round(
                rng.normal(loc=pdf_views * 0.8 + video_views * 0.6, scale=1), 1
            )
            student[f"quiz_avg_medium_1_{i}"] = round(
                rng.normal(loc=pdf_views * 0.6 + video_views * 0.7, scale=1.2), 1
            )
            student[f"quiz_avg_hard_1_{i}"] = round(
                rng.normal(loc=pdf_views * 0.4 + video_views * 0.8, scale=1.5), 1
            )

        # C√°c h√†nh vi kh√°c
        student["forum_post_count"] = rng.integers(0, 10)
        student["assignment_status"] = rng.integers(0, 2)  # 0 ho·∫∑c 1

        data.append(student)

    df = pd.DataFrame(data)

    # L√†m s·∫°ch d·ªØ li·ªáu (c·∫Øt ƒëi·ªÉm v·ªÅ [0, 10])
    for col in df.columns:
        if col.startswith("quiz_avg_"):
            df[col] = df[col].clip(0, 10)

    print("üìä 5 d√≤ng d·ªØ li·ªáu ƒë·∫ßu:")
    print(df.head())

    print("\nüîç Th·ªëng k√™:")
    print(df.describe())

    return df
```

```python
df = generate_student_log_data(n_samples=5000, random_state=42)
# L∆∞u file n·∫øu c·∫ßn:
df.to_csv("data/student_logs_chapter1_full.csv", index=False)
```
